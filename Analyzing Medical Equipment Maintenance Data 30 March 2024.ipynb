# Another use case involving medical equipment could be analyzing equipment maintenance data to ensure optimal performance and minimize downtime. Let's consider a scenario where a hospital wants to analyze maintenance records for their medical equipment to identify patterns, predict maintenance needs, and optimize equipment uptime.

# Use Case: Analyzing Medical Equipment Maintenance Data
Suppose we have a dataset containing maintenance records for various medical equipment used in a hospital. Each record includes information such as equipment type, maintenance date, maintenance type (preventive or corrective), and duration of maintenance. We want to analyze this data to:

Identify equipment with high maintenance frequency.

Determine the most common types of maintenance performed.

Visualize maintenance trends over time.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample maintenance data (can be loaded from a file or database)
data = {
    'Equipment': ['MRI', 'X-Ray', 'Ultrasound', 'MRI', 'CT Scanner', 'X-Ray', 'Ultrasound'],
    'Maintenance_Date': ['2023-01-01', '2023-01-05', '2023-01-10', '2023-01-15', '2023-01-20', '2023-01-25', '2023-01-30'],
    'Maintenance_Type': ['Preventive', 'Corrective', 'Preventive', 'Corrective', 'Preventive', 'Corrective', 'Preventive'],
    'Duration_hours': [2, 4, 1, 3, 2, 5, 1]
}
data
```

```python
# Convert data to DataFrame
df = pd.DataFrame(data)
df
```

```python
# Convert 'Maintenance_Date' column to datetime
df['Maintenance_Date'] = pd.to_datetime(df['Maintenance_Date'])
df['Maintenance_Date']
```

```python
# 1. Identify equipment with high maintenance frequency
maintenance_frequency = df['Equipment'].value_counts()
maintenance_frequency
```

```python
# Plot equipment with high maintenance frequency
plt.figure(figsize=(10, 6))
maintenance_frequency.plot(kind='bar', color='lightgreen')
plt.title('Equipment with High Maintenance Frequency')
plt.xlabel('Equipment')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 2. Determine the most common types of maintenance performed
maintenance_types = df['Maintenance_Type'].value_counts()
```

```python
# Plot the most common types of maintenance performed
plt.figure(figsize=(6, 4))
maintenance_types.plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightcoral'], explode=(0.1, 0), labels=None)
plt.title('Most Common Types of Maintenance')
plt.axis('equal')
plt.legend(labels=maintenance_types.index, loc='upper right')
plt.tight_layout()
plt.show()
```

```python
# 3. Visualize maintenance trends over time
plt.figure(figsize=(10, 6))
for equipment, group in df.groupby('Equipment'):
    plt.plot(group['Maintenance_Date'], group['Duration_hours'], marker='o', label=equipment)
plt.title('Maintenance Trends Over Time')
plt.xlabel('Maintenance Date')
plt.ylabel('Duration (hours)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
