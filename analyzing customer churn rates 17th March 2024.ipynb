# Problem Statement:

You work for a telecommunications company and are tasked with analyzing customer churn rates to identify factors influencing customer retention.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Hypothetical dataset containing customer churn data
data = {
    'Customer_ID': range(1, 101),
    'Age': [25, 30, 40, 35, 50] * 20,
    'Gender': ['Male', 'Female'] * 50,
    'Location': ['Urban', 'Suburban', 'Rural'] * 33 + ['Urban'],
    'Data_Usage': [5, 10, 15, 8, 12] * 20,
    'Calls_Made': [20, 30, 25, 35, 15] * 20,
    'Customer_Support_Satisfaction': [4, 3, 5, 4, 2] * 20,
    'Churn_Status': ['Yes', 'No'] * 50
}
```

```python
data
```

```python
# Create a DataFrame from the data
df = pd.DataFrame(data)
df
```

```python
# Visualization 1: Churn Rate by Customer Demographics (Bar Plot)
plt.figure(figsize=(8, 6))
sns.countplot(x='Age', hue='Churn_Status', data=df)
plt.title('Churn Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Customers')
plt.legend(title='Churn Status', labels=['Churned', 'Retained'])
plt.show()
```

```python
# Visualization 2: Churn Rate by Service Usage (Violin Plot)
plt.figure(figsize=(8, 6))
sns.violinplot(x='Churn_Status', y='Data_Usage', data=df)
plt.title('Churn Rate by Data Usage')
plt.xlabel('Churn Status')
plt.ylabel('Data Usage')
plt.show()
```

```python
# Visualization 3: Impact of Customer Support (Box Plot)
plt.figure(figsize=(8, 6))
sns.boxplot(x='Customer_Support_Satisfaction', y='Churn_Status', data=df)
plt.title('Impact of Customer Support on Churn Rate')
plt.xlabel('Customer Support Satisfaction')
plt.ylabel('Churn Status')
plt.show()
```
