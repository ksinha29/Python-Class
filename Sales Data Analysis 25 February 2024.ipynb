# Problem Statement: Sales Data Analysis

Problem Description:
You are given a dataset containing information about daily sales for a business over a certain period. The dataset includes the following columns: Date, Product, Order_Quantity... and Revenu . Your task is to perform basic statistical analysis to gain insights into the sales performance.

Tasks to Perform:

Load the Dataset:

Load the sales dataset from a CSV file or any other suitable format.
Data Cleaning:

Check for missing values and handle them appropriately.
Ensure data types are correct.
Descriptive Statistics:

Calculate and display descriptive statistics for key metrics (e.g., Order_Quantity, Revenue).
Provide insights into the average daily sales, total revenue, etc.
Time Series Analysis:

Analyze the sales trends over time. Plot a time series graph to visualize the changes in daily sales.
Product Performance:

Identify the top-selling products based on the total Order_Quantity or revenue.
Provide insights into the product performance.
Correlation Analysis:

Explore the correlation between different variables (e.g.,Order_Quantity and Revenue).
Determine if there are any significant relationships.
Monthly Aggregation:

Aggregate the data on a monthly basis and analyze the monthly trends.
Visualizations:

Create visualizations such as bar charts, line plots, or pie charts to present key findings.
Sample Code:

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Task 1: Load the Dataset
sales_data = pd.read_csv('sales_data.csv')  # Replace 'sales_data.csv' with your actual file path
sales_data
```

```python
# Task 2: Data Cleaning
sales_data.dropna(inplace=True)
sales_data['Date'] = pd.to_datetime(sales_data['Date'])
sales_data
```

```python
# Task 3: Descriptive Statistics
print("Descriptive Statistics:")
print(sales_data[['Order_Quantity', 'Revenue']].describe())
```

```python
# Task 4: Time Series Analysis
plt.figure(figsize=(10, 5))
plt.plot(sales_data['Date'], sales_data['Order_Quantity'], label='Units Sold')
plt.plot(sales_data['Date'], sales_data['Revenue'], label='Revenue')
plt.title('Daily Sales Trends')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
# Task 5: Product Performance
top_products = sales_data.groupby('Product')['Order_Quantity'].sum().nlargest(5)
print("\nTop Selling Products:")
print(top_products)
```

```python
# Task 6: Correlation Analysis
correlation_matrix = sales_data[['Order_Quantity', 'Revenue']].corr()
print("\nCorrelation Matrix:")
print(correlation_matrix)
```

```python
# Task 7: Monthly Aggregation
monthly_sales = sales_data.resample('M', on='Date').sum()
```

```python
monthly_sales
```

```python
# Task 8: Visualizations
plt.figure(figsize=(12, 6))
plt.bar(monthly_sales.index, monthly_sales['Revenue'], color='blue', alpha=0.7, label='Monthly Revenue')
plt.title('Monthly Sales Trends')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.legend()
plt.show()
```
