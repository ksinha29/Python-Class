# Problem Statement:
Suppose you work for a transportation company that provides ride-sharing services, and you're tasked with analyzing customer feedback to identify areas for improvement in service quality.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Hypothetical dataset containing customer reviews and ratings
data = {
    'Review': ['Clean vehicle', 'Friendly driver', 'Late arrival', 'Clean vehicle', 'Late arrival'],
    'Rating': [5, 4, 2, 5, 3]
}
```

```python
# Create a DataFrame from the data
df = pd.DataFrame(data)
```

```python
# Plotting a box plot to visualize the relationship between vehicle cleanliness and ratings
plt.figure(figsize=(8, 6))
sns.boxplot(x='Review', y='Rating', data=df)
plt.title('Impact of Vehicle Cleanliness on Customer Ratings')
plt.xlabel('Customer Review')
plt.ylabel('Rating')
plt.show()
```

# Analysis:
To address this problem, you decide to analyze customer reviews and ratings. You have access to a dataset containing customer reviews along with corresponding ratings on a scale of 1 to 5.

# Conclusion:
After analyzing the data, you discover that there's a noticeable correlation between the cleanliness of vehicles and customer satisfaction. You hypothesize that maintaining cleaner vehicles could lead to higher customer ratings and improved service quality.
