```python
import matplotlib.pyplot as plt
survival_by_class = data2.groupby('Country/Region')['Recovered'].count()
survival_by_class.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Got saved')
plt.xlabel('might get saved')
plt.ylabel('actually got saved')
plt.show()
print(survival_by_class)
```

```python
plt.hist(data2['Active'], bins=20, color='skyblue', edgecolor='black')
plt.title('title1')
plt.xlabel('title2')
plt.ylabel('title3')
plt.show()
```

```python
import seaborn as sns
```

```python
numeric_columns = data2.select_dtypes(include=['float64', 'int64']).columns


correlation_matrix = data2[numeric_columns].corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.countplot(x='WHO Region', data=data2)
plt.title('Distribution')
plt.show()
```

```python
sns.barplot(x='Recovered', y='Recovered', data=data2)
plt.title('Recovery')
plt.show()
```
