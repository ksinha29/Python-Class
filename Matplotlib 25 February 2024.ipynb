# Matplotlib to create different types of graphs: a line plot, a bar plot, a scatter plot, and a pie chart.

```python
#pip install matplotlib
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generate data
x_values = np.linspace(0, 10, 100)
y_values = np.sin(x_values)
```

```python
y_values
```

```python
x_values
```

```python
categories = ['Category A', 'Category B', 'Category C']
values = [4, 7, 5]
```

```python
categories
```

```python
values
```

```python
# Line plot
plt.figure(figsize=(10, 5))
plt.subplot(2, 2, 1)
plt.plot(x_values, y_values, label='Sin Function')
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
```

```python
# Bar plot
plt.subplot(2, 2, 2)
plt.bar(categories, values, color=['blue', 'orange', 'green'])
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
```

```python
# Scatter plot
plt.subplot(2, 2, 3)
x_scatter = np.random.rand(20)
y_scatter = np.random.rand(20)
plt.scatter(x_scatter, y_scatter, color='red', marker='o')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

```python
# Pie chart
plt.subplot(2, 2, 4)
plt.pie(values, labels=categories, autopct='%1.1f%%', colors=['gold', 'lightcoral', 'lightskyblue'])
plt.title('Pie Chart')
```

```python
# Adjust layout
plt.tight_layout()
```

```python
# Show the plots
plt.show()
```
